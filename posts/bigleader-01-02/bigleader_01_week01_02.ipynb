### 결측치 처리
결측치 처리에 관해선 정답이 없다는게 내 주관적인 생각.  
맞다고 생각되면 일단 해보는걸 추천함.

```python
# Age : 177 / Cabin : 687 / Embarked : 2
df.isnull().sum()
```

결측치를 무작정 제거하는건 좋은 생각이 아님

```python
#df.dropna(subset=['Age', 'Cabin'])
```

탑승객들의 나이는 어떤식으로 결측치 처리 할 것인가

```python
# 그냥 평균으로 넣을 수도 있음.

# df['Age'].fillna(df['Age'].mean())
```

```python
# 생존자 나이 평균
mean1 = df[df['Survived'] == 1]['Age'].mean()

# 사망자 나이 평균
mean0 = df[df['Survived'] == 0]['Age'].mean()

print(mean1, mean0)
```

```python
df.loc[df['Survived'] == 1, 'Age']
```

```python
df[df['Survived'] == 1]['Age'] # 생존그룹에서 나이만 추출
```

```python
# 앞에는 생존자 그룹을 분류 / 뒤에는
df.loc[df['Survived'] == 1, 'Age'] = df[df['Survived'] == 1]['Age'].fillna(mean1)
df.loc[df['Survived'] == 0, 'Age'] = df[df['Survived'] == 0]['Age'].fillna(mean0)
```
